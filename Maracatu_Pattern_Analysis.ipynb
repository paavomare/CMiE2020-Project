# Rhythmic Pattern Statistics in the Maracatu Renascente Rhythm Corpus

## Introduction
My project consists of an analysis of a Brazilian style of percussion music called *aracatu de baque virado*. Characteristically, many accented notes of the rhythms of the bass drums *alfaias* are not played on the beat (as compared to *samba* for example) but rather on the second or third sixteenth note of the beat. This is one of the crucial features of the music. During this small project I want to find out what the ratio between on-beat and off-beat notes is.

First, I need to compute how many accented notes of the most common *maracatu* rhythms are played on-beat and off-beat by checking their offset values and their articulations via Music21. For that I use MuseScore files with the transcribed rhythms, which are used by my own band Maracatu Renascente. Apart from *maracatu*, our repertoire also comprises other styles, for example *coco*, *afoxÃ©*, and *mangue*. Instead of strictly focussing on *maracatu* rhythms, I will also include those patterns.

Secondly, by utilizing the information from the first step, the program will produce pie charts for each rhythmic pattern representing the ratio of each found element and another pie chart for the total ratio of all elements in all rhythmic patterns.

# Central Questions
The main goal of this project is to write a code that produces information about the following issues for every pattern in the Maracatu Renascente rhythm corpus:
* how many notes in each pattern and in total are on-beat?
* how many notes in each pattern and in total are off-beat?
* how many accented notes in each pattern and in total are on-beat?
* how many accented notes in each pattern and in total are off-beat?
* how many rests are found in the patterns?

# Code
The first thing to do is to import certain Python modules that are needed to use functions for music analysis (`music21`), handling the data (`os`) and finally plotting graphs (`matplotlib.pyplot`).

When using the following code, it is essential to put the data, that is the .musicxml-files, in the same directory as in the `folder` variable. As the directory comprises also .pdf- and .mscz-files, the program puts all the desired .musicxml-files into a separate list (`xmls`) by separating all the files whose file names end with 'xml'.

```python
from music21 import *
import os
import matplotlib.pyplot as plt

folder = "./maracatu_patterns/"

allFiles = os.listdir(folder)

xmls = []

for f in allFiles:
    if f[-3:] == 'xml':
        xmls.append(f)
```

In the following cell, the user can adjust the settings of the display of extra information in the code. 

To get more information about the process and for a easier and comparison between the code output and the score as PDF that may be desired from the user, the program can print sentences for each note or rest via the function `print_comment()`, so that the user can check if the notation and the output are identical or if there are mistakes. Currently, those auxiliary sentences are not shown. If the user sets the variable `print_aux_comments` from the default value `False` to `True`, the program will print the comments accordingly. 

```python
def print_comment(comment):
    """
    If the variable print_aux_comments is set to True, 
    this function prints auxiliary comments in the statistics output, that is:
    1) Information about each found element and its characteristics (on-beat, off-beat, accented, not accented, rest)
    2) List of percentages of each occuring element in each rhythm
    
    Args:
        comment (str): the given comment that contains auxiliary information
    
    Returns:
        comment (str): the auxiliary comment
        
    >>> print_aux_comments is True
    >>> print_comment("Hello World!")
    Hello World!
    
    >>> print_aux_comments is False
    >>> print_comment("Hello World!")
    
    """
    print_aux_comments is False
    if print_aux_comments is True:
        print(comment)

# Set this variable to True for auxiliary sentences being printed
print_aux_comments = False
```

To compute the percentage of the occurrence of each element in the final steps, six counters need to be defined for the different categories: 
* on-beat notes 
* off-beat notes
* accented on-beat notes
* accented off-beat notes
* rests
* all elements.

```python
on_beat_counter = 0
off_beat_counter = 0
acc_on_beat_counter = 0
acc_off_beat_counter = 0
rest_counter = 0
elements_counter = 0
```

For each element in the stream the program checks if it is a note or a rest. If it is a note another `if`-statement checks if the note is positioned on-beat or off-beat. In yet another `if`-statement the program checks if the note is accented or not. Finally, the counters for each possible phenomenon is raised by 1 if the very element occurs.

For each pattern, the program computes the occurrence of each element as a percentage.

Before showing the pie chart, the occurence is presented in absolute values.

The code for the pie chart is mostly copied from the basic pie chart 
of the matplotlib website: 
(https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py).

The labels are used for the legend rather than putting them into the pie chart, which 
produces a cleaner diagram. If an element does not occur in pattern at all, a lambda 
function in `autopct` removes the 0.0%-value from the pie chart to ensure a better readability. The title of each pie chart is set by using all letters of the file name from the first position (0) to the nineth when counting from the end of the file name (-9), because '.musicxml' is nine characters long and needs to be removed by slicing.

Optional: `print_aux_comments` can be set to `True` for further information during the processing. 

```python
for xml in xmls:
    s = converter.parse(folder + xml)
    print("Processing", xml)
    print()
    nr = s.flat.notesAndRests.stream()
    pattern_on_beat_counter = 0
    pattern_off_beat_counter = 0
    pattern_acc_on_beat_counter = 0
    pattern_acc_off_beat_counter = 0
    pattern_rest_counter = 0
    pattern_elements_counter = 0
    
    for n in nr:
        beat = n.beat
        elements_counter += 1
        pattern_elements_counter += 1
        if n.isNote:
            if beat % 1 == 0:
                if n.articulations and n.articulations[0].name == 'accent':
                    print_comment("\t {} has an accent.".format(n))
                    acc_on_beat_counter += 1
                    pattern_acc_on_beat_counter += 1
                else:
                    print_comment("\t {} has no accent.".format(n))   
                    on_beat_counter += 1
                    pattern_on_beat_counter += 1 
            elif beat % 1 > 0:
                if n.articulations and n.articulations[0].name == 'accent':
                    print_comment("\t {} has an accent.".format(n))
                    acc_off_beat_counter += 1
                    pattern_acc_off_beat_counter += 1
                else:
                    print_comment("\t {} has no accent.".format(n))
                    off_beat_counter += 1
                    pattern_off_beat_counter += 1
        elif n.isRest:
            print_comment("\t {} is a rest.".format(n))
            rest_counter += 1
            pattern_rest_counter += 1
        
        # For each pattern, the program computes the occurrence of each element as a percentage.
        pattern_percentage_on_beat = pattern_on_beat_counter / pattern_elements_counter * 100
        pattern_percentage_acc_on_beat = pattern_acc_on_beat_counter / pattern_elements_counter * 100
        pattern_percentage_off_beat = pattern_off_beat_counter / pattern_elements_counter * 100
        pattern_percentage_acc_off_beat = pattern_acc_off_beat_counter / pattern_elements_counter * 100
        pattern_percentage_rests = pattern_rest_counter / pattern_elements_counter * 100
    
    print("Notes on-beat:", pattern_on_beat_counter)
    print("Notes off-beat:", pattern_off_beat_counter)
    print("Accented notes on-beat:", pattern_acc_on_beat_counter)
    print("Accented notes off-beat:", pattern_acc_off_beat_counter)
    print("Rests:", pattern_rest_counter)
    print()
    print_comment("On-beat: {:.1f}%".format(pattern_percentage_on_beat))
    print_comment("Accented On-beat: {:.1f}%".format(pattern_percentage_acc_on_beat))
    print_comment("Off-beat: {:.1f}%".format(pattern_percentage_off_beat))
    print_comment("Accented Off-beat: {:.1f}%".format(pattern_percentage_acc_off_beat))
    print_comment("Rests: {:.1f}%".format(pattern_percentage_rests))
    
    # Code for the pie chart
    labels = ['On-beat', 'Accented on-beat', 'Off-beat', 'Accented off-beat', 'Rests']
    sizes = [pattern_percentage_on_beat, pattern_percentage_acc_on_beat, pattern_percentage_off_beat, 
             pattern_percentage_acc_off_beat, pattern_percentage_rests]
            
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '', startangle=90)
    
    ax1.legend(labels, title = 'Legend', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    ax1.set_title(xml[0:-9])
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.show()
```

The program now computes the final results, i.e. the occurrence of each element in all rhythmic patterns as a percentage.    

```python
percentage_on_beat = on_beat_counter / elements_counter * 100
percentage_acc_on_beat = acc_on_beat_counter / elements_counter * 100
percentage_off_beat = off_beat_counter / elements_counter * 100
percentage_acc_off_beat = acc_off_beat_counter / elements_counter * 100
percentage_rests = rest_counter / elements_counter * 100
```

Lastly, the program presents the final results and the pie chart. 

The code for the last pie chart is basically the same as the one before, only `ax1.set_title` differs as now it should not use part of the file name as title but rather just "Final Results".

```python
print("Final Results")
print("On-beat:", on_beat_counter)
print("Accented on-beat:", acc_on_beat_counter)
print("Off-beat:", off_beat_counter)
print("Accented off-beat:", acc_off_beat_counter)
print("Rests:", rest_counter)
print()
print("On-beat: {:.1f}%".format(percentage_on_beat))
print("Accented On-beat: {:.1f}%".format(percentage_acc_on_beat))
print("Off-beat: {:.1f}%".format(percentage_off_beat))
print("Accented Off-beat: {:.1f}%".format(percentage_acc_off_beat))
print("Rests: {:.1f}%".format(percentage_rests))

# Code for the pie chart
labels = ['On-beat', 'Accented on-beat', 'Off-beat', 'Accented off-beat', 'Rests']
sizes = [percentage_on_beat, percentage_acc_on_beat, percentage_off_beat, percentage_acc_off_beat, percentage_rests]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '', startangle=90)

ax1.legend(labels, title = 'Legend', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax1.set_title("Final Results")
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
